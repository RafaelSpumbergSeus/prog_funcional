## Exercícios com Programação Funcional: Listas Encadeadas

```python
def head(ll):
    return ll[0]

def tail(ll):
    return ll[1]
```

```python
# Exemplo:
ll = (1, (2, (3, None)))
print(head(ll))
print(tail(ll))
```

### 1. Crie uma função que adicione um item no início de lista encadeada

```python
def add(ll, elem):
    return (elem, ll)

print(add(ll, 0))
```

### 2. Crie uma função que adicione um item no final de lista encadeada

```python
# def add_end(ll, elem):
#     if ll[1] == None:
#         return (ll[0], (elem, None))
#     else:
#         return (ll[0], add_end(ll[1], elem))

def add_end(elem, ll):
    if not ll:
        return (elem, None)
    return (head(ll), add_end(elem, tail(ll)))
    
print(add_end(4, ll))
```

### 3. Transformando uma lista em uma lista encadeada (linked)

```python
lista = [1, 2, 3, 4, 5]

def transformar_em_ll(lista):
    if not lista:
        return None
    return (lista[0], transformar_em_ll(lista[1:]))

print(transformar_em_ll(lista))
```

### 4. Transformando uma lista encadeada (linked) em lista

```python
def transformar_ll_em_lista(ll):
    if not ll:
        return []
    return [head(ll)] + transformar_ll_em_lista(tail(ll))
    # return [ll[0]] + transformar_ll_em_lista(ll[1:])
    
print(ll)
print(transformar_ll_em_lista(ll))
```

### 5. Função que retorna a quantidade de elementos

```python
def count(ll):
    if not ll:
        return 0
    return 1 + count(tail(ll))

print(count(ll))
```

### 6. Crie uma função que retorne o último elemento de uma lista encadeada

```python
def return_last(ll):
    if tail(ll) == None:
        return head(ll)
    return return_last(tail(ll))

print(return_last())
```

### 5. Crie uma função de alta ordem que receba uma função e uma lista encadeada, e aplique a função para todos os objetos da lista encadeada


```python
def apply(f, ll):
    if tail(ll) == None:
        return (f(head(ll)), None)
    return (f(head(ll)), apply(f, tail(ll)))

print(apply(lambda x: x**2, transformar_em_ll([1,2,3,4,5,6,7,8])))
```

### 6. Crie uma função que retorne o maior elemento de uma lista encadeada

```python
def maior(ll):
    if tail(ll) == None:
        return head(ll)
    return max(head(ll), maior(tail(ll)))

print(maior(ll))
```

### 7. Crie uma função que retorne uma lista encadeada contendo apenas os elementos pares de uma lista encadeada

```python
def filter_pares(ll):
    if tail(ll) == None:
        if head(ll) % 2 == 0:
            return ll
        return None
    
    if head(ll) % 2 == 0:
        return (head(ll), filter_pares(tail(ll)))
    return filter_pares(tail(ll))

print(filter_pares(transformar_em_ll([1,2,3,4,5,6,7,8])))
```

### 8. Crie uma função que leva o primeiro item da lista para a última posição

```python
def send_first_to_end(ll):
    """ send the original first element to the end of the list """
    return add_end(tail(ll), head(ll))

print(send_first_to_end(transformar_em_ll([1,2,3,4,5,6,7,8])))
```

### 9. Crie uma função que inverta a lista encadeada

```python
def invert_list(ll):
    if tail(ll) == None:
        return ll
    return add_end(invert_list(tail(ll)), head(ll))

print(invert_list(transformar_em_ll([1,2,3,4,5,6,7,8])))
```

### 9. MergeSort para Lista Encadeadas

#### A. Função Split

Escreva uma função pura `split`, que receba uma lista `L` e retorne duas listas (`L1` e `L2`) tal que as seguinte propriedades sejam verdadeiras.
- $elem$ in `L` $->$  $elem$ in `L1` $||$ `L2`
- `size(L)` $==$ `size(L1)` $+$ `size(L2)`  (não pode ter duplicação)
- `abs(size(L1)` - `size(L2))` $<=$ `1`

```python
def split(linked_list):

    def divide(lst, n):
        if n == 0 or lst is None:
            return None, lst
        first_half, second_half = divide(lst[1], n - 1)
        return (lst[0], first_half), second_half
    
    return divide(linked_list, count(linked_list) // 2)

print(split(transformar_em_ll([1,2,3,4,5,6,7,8])))
```

#### B. Função sorted em ascendente

Escrever uma função pura `sorted`, que retorne `True` quando `L` for uma lista ordenada em ordem ascendente e `False` caso contrário.

```python
def sorted(ll):
    if tail(ll) == None:
        return True
    return head(ll) <= head(tail(ll)) and sorted(tail(ll))

print(sorted(transformar_em_ll([1,2,3,4,5,6,7,8])))
print(sorted(transformar_em_ll([1,2,3,4,5,7, 6,8])))
```

#### C. Função merge
 Escreva uma função pura `merge`, que receba duas listas (`L1`, `L2`) e retorne uma lista `L`, tal que as seguintes proprieades sejam verdadeiras.
 
- $elem$ in `L1` $||$ `L2` $->$ $elem$ in `L` 
- `size(L1)` $+$ `size(L2)` $==$ `size(L)` 
- `sorted(L1)` $&&$ `sorted(L2))` $->$ `L`

```python
def merge(ll1, ll2):
    if ll1 == None:
        return ll2
    if ll2 == None:
        return ll1
    if head(ll1) < head(ll2):
        return (head(ll1), merge(tail(ll1), ll2))
    return (head(ll2), merge(ll1, tail(ll2)))

print(merge(transformar_em_ll([1,3,5,7,9]), transformar_em_ll([2,4,6,8,10])))
```

#### D. Função mergeSort

Escreva uma função pura `mergeSort`, que receba uma lista `L` e retorne uma lista ordenada `LS`, tal que as seguintes propriedades sejam verdadeiras

- $elem$ in `L` $<->$ $elem$ in `LL` (preserva os elementos)
- `sorted(LL)`


```python
def merge_sort(ll):
    if tail(ll) == None:
        return ll
    first_half, second_half = split(ll)
    return merge(merge_sort(first_half), merge_sort(second_half))

print(merge_sort(transformar_em_ll([1,3,5,7,9,2,4,6,8,10])))
```
